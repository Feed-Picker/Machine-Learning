```python
#contoh penerapan
stemmer = PorterStemmer()
input_str = "ada berapa banyak kasus yang ingin kamu selesaikan hari ini?"
input_str = nltk.word_tokenize(input_str)
for word in input_str:
    print(stemmer.stem(word))
```

```python
lemmatizer = WordNetLemmatizer()
for index, row in datasetku.iterrows():
    filter_sentence = []
    sentence = row["Review"]
    sentence = re.sub(r'[^\w\s]','',sentence)
    sentence = sentence.lower()
    words = nltk.word_tokenize(sentence)
    words = [w for w in words if not w in stop_words]
    for word in words:
        filter_sentence.append(lemmatizer.lemmatize(word))
    print(filter_sentence)
    datasetku.at[index, 'Review'] = filter_sentence
```

```python
datasetku = datasetku.dropna(axis=1, how='all')
datasetku
```
